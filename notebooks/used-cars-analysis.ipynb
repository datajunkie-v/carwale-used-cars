```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/content/cars1.csv')
df2 = pd.read_csv('/content/cars2.csv')
df3 = pd.read_csv('/content/cars3.csv')
df4 = pd.read_csv('/content/cars4.csv')
df5 = pd.read_csv('/content/cars5.csv')
df6 = pd.read_csv('/content/cars6.csv')
df7 = pd.read_csv('/content/cars7.csv')
df8 = pd.read_csv('/content/cars8.csv')
df9 = pd.read_csv('/content/cars9.csv')
df10 = pd.read_csv('/content/cars10.csv')
```

```python
all_dfs = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10]
```

```python
cars_df = pd.concat(all_dfs,ignore_index=True)
```

```python
cars_df
```

```python
cars_df.info()
```

## **Data Cleaning**

```python
cars_df['year'] = cars_df['title'].str[:4]
```

```python
cars_df
```

```python
#converting string numbers to integers

cars_df['title'] = cars_df['title'].str[5:]
cars_df['prices'] = cars_df['prices'].str[4:]
cars_df['prices'] = cars_df['prices'].str.strip('Lakh')
cars_df['prices'] = cars_df['prices'].str.strip('Crore')
cars_df['prices'] = cars_df['prices'].str.replace(',', '')
```

```python
cars_df
```

```python
#Multiplying Lakh with 100000

cars_df['prices'] = cars_df['prices'].apply(float)
cars_df['prices'] = cars_df['prices'].multiply(100000)
```

```python
#Splitting combined columns into different columns

cars_df[['distance_travelled','Fuel','Location']] = cars_df['distance'].str.split("|",n=2,expand = True)
```

```python
cars_df.drop(['distance'],axis = 1,inplace = True)
```

```python
cars_df.rename(columns={"distance_travelled": "distance(kms)"},inplace = True)
```

```python
cars_df['distance(kms)']
```

```python
cars_df['distance(kms)'] = cars_df['distance(kms)'].str.replace(',', '')
cars_df['distance(kms)'] = cars_df['distance(kms)'].str.replace(' ', '')
```

```python
cars_df['distance(kms)'] = cars_df['distance(kms)'].str[:-4]
```

```python
cars_df['distance(kms)'] = cars_df['distance(kms)'].apply(int)
```

```python
cars_df['Brand'] = cars_df['title'].str.split(" ",n=1).str[0]
```

```python
cars_df
```

```python
cars_df.info()
```

```python
#dropping brands with less than 25 cars

cars_df = cars_df[cars_df["Brand"].str.contains("Volvo|Jaguar|Nissan|Porsche|Datsun|MINI|Isuzu|Lexus|Chevrolet|Bentley|Fiat|Mitsubishi|Aston|Lamborghini|Rolls-Royce") == False]
```

```python
import seaborn as sns

sns.scatterplot(cars_df, x='distance(kms)', y='prices')
```

```python
#deleting outliers

q = cars_df["prices"].quantile(0.99)
cars_df = cars_df[cars_df["prices"] < q]
```

```python
q = cars_df["distance(kms)"].quantile(0.99)
cars_df = cars_df[cars_df["distance(kms)"] < q]
```
